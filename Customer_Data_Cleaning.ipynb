**Libraries import + load data**

```python
import pandas as pd

df = pd.read_csv("olist_customers_dataset.csv")

df.shape
```

**Dataset preview**

```python
df.head()
```

**Missing values check**

```python
df.isnull().sum()
```

**Duplicate rows remove**

```python
df = df.drop_duplicates()
df.shape
```

**Handle missing values**

```python
df['customer_city'] = df['customer_city'].fillna("Unknown")
df['customer_state'] = df['customer_state'].fillna("Unknown")
```

**Text cleaning**

```python
df['customer_city'] = df['customer_city'].str.lower().str.strip()
df['customer_state'] = df['customer_state'].str.upper().str.strip()
```

**Zip code validation**

```python
df = df[df['customer_zip_code_prefix'] > 0]
df.shape
```

**Final check**


```python
df.info()
```

**Save cleaned file**

```python
df.to_csv("cleaned_customers.csv", index=False)
```
